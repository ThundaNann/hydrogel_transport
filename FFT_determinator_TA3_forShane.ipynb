```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import tkinter as tk
from tkinter import filedialog, messagebox
import os
from pathlib import Path
from datetime import datetime

# Set default font sizes for publication quality - LARGE
plt.rcParams.update({
    'font.size': 20,
    'axes.titlesize': 24,
    'axes.labelsize': 22,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
    'legend.fontsize': 18,
    'figure.titlesize': 26,
    'axes.linewidth': 1,
    'grid.linewidth': 1,
    'xtick.major.width': 1.5,
    'ytick.major.width': 1.5,
    'xtick.major.size': 8,
    'ytick.major.size': 8,
    'xtick.minor.visible': True,
    'ytick.minor.visible': True,
    'xtick.minor.size': 4,
    'ytick.minor.size': 4,
    'xtick.minor.width': 1,
    'font.weight': 'normal',  # Unbold fonts
    'axes.labelweight': 'normal'  # Unbold axis labels
})

def select_files_or_folder():
    """
    Open dialog to select either CSV files or a folder for recursive search
    Returns: (file_paths, base_folder) tuple
    """
    root = tk.Tk()
    root.withdraw()
    root.lift()
    root.attributes('-topmost', True)
    
    # Ask user for selection mode
    choice = messagebox.askyesnocancel(
        "Selection Mode",
        "YES: Select a folder (recursive search for CH1.CSV files)\n"
        "NO: Select individual CSV files\n"
        "CANCEL: Exit"
    )
    
    if choice is None:  # Cancel
        return [], None
    elif choice:  # Yes - folder selection
        folder_path = filedialog.askdirectory(title="Select folder to search for CH1.CSV files")
        if folder_path:
            # Recursively find all CSV files in folders and subfolders
            # NOTE: This does NOT look inside zip/compressed files, only in directory structure
            csv_files = set()  # Use set to avoid duplicates
            
            # Search for both .csv and .CSV extensions
            for pattern in ['*.csv', '*.CSV']:
                for path in Path(folder_path).rglob(pattern):
                    if path.is_file():
                        # Only include CH1 files (filter out RFA and others)
                        if 'CH1' in path.name.upper():
                            csv_files.add(str(path))
            
            # Convert set back to sorted list
            csv_files = sorted(list(csv_files))
            
            if csv_files:
                print(f"\nFound {len(csv_files)} CH1.CSV files in folder and subfolders:")
                for f in csv_files:
                    print(f"  - {os.path.relpath(f, folder_path)}")
                
                # Ask if user wants to proceed with all files
                proceed = messagebox.askyesno(
                    "Confirm",
                    f"Found {len(csv_files)} CH1.CSV files. Process all?"
                )
                if proceed:
                    return csv_files, folder_path  # Return both files and the selected folder
            else:
                messagebox.showinfo("No files found", "No CH1.CSV files found in the selected folder.")
                return [], None
    else:  # No - individual file selection
        # Note: Manual selection allows any CSV file, not just CH1
        file_paths = filedialog.askopenfilenames(
            title="Select CSV files (can select any CSV, not just CH1)",
            filetypes=[("CSV files", "*.csv"), ("CSV files", "*.CSV"), ("All files", "*.*")]
        )
        if file_paths:
            # For individual files, use the parent directory of the first file
            base_folder = os.path.dirname(file_paths[0])
            return file_paths, base_folder
        return [], None
    
    return [], None

def read_oscilloscope_csv(file_path):
    """
    Read oscilloscope CSV data
    Adjust skiprows and column indices based on your CSV format
    """
    # Read CSV - adjust skiprows based on your file header
    data = pd.read_csv(file_path, skiprows=15)
    
    # Adjust these column indices based on your CSV structure
    time_col = data.columns[3]  # Usually column 3 is time
    voltage_col = data.columns[4]  # Usually column 4 is voltage
    
    # Skip first few rows if they contain non-data
    time_data = data[time_col][3:].astype(float)
    voltage_data = data[voltage_col][3:].astype(float)
    
    return time_data, voltage_data

def perform_fft_analysis(time_data, voltage_data, start_time=30e-6, end_time=90e-6):
    """
    Perform FFT on specified time window and return both linear and dB scales
    
    Parameters:
    time_data: array of time values (in seconds)
    voltage_data: array of voltage values
    start_time: start of analysis window (default 30 s)
    end_time: end of analysis window (default 90 s)
    """
    # Select data within time window
    time_mask = (time_data >= start_time) & (time_data <= end_time)
    selected_time = time_data[time_mask]
    selected_voltage = voltage_data[time_mask]
    
    # Calculate sampling rate
    dt = np.mean(np.diff(selected_time))
    sampling_rate = 1 / dt
    
    # Perform FFT
    n = len(selected_voltage)
    frequencies = np.fft.fftfreq(n, dt)
    fft_values = np.fft.fft(selected_voltage)
    
    # Calculate AMPLITUDE spectrum (not magnitude)
    amplitude_spectrum = np.abs(fft_values) * (2.0 / n)  # Scale for amplitude
    amplitude_spectrum[0] = amplitude_spectrum[0] / 2  # Correct DC component if present
    
    # Calculate both linear and dB scales
    amplitude_linear = amplitude_spectrum.copy()
    amplitude_db = 20 * np.log10(amplitude_spectrum + 1e-10)  # Add small value to avoid log(0)
    
    # Get positive frequencies only (up to Nyquist)
    positive_freq_mask = frequencies > 0
    pos_frequencies = frequencies[positive_freq_mask]
    pos_amplitude_linear = amplitude_linear[positive_freq_mask]
    pos_amplitude_db = amplitude_db[positive_freq_mask]
    
    return pos_frequencies, pos_amplitude_linear, pos_amplitude_db, selected_time, selected_voltage, sampling_rate

def plot_results_both_scales(time_data, voltage_data, frequencies, amplitude_linear, amplitude_db,
                            selected_time, selected_voltage, file_path, sampling_rate, show_plots=True, plots_folder=None):
    """
    Create plots showing full waveform, selected window, and FFT in BOTH linear and dB scales
    
    Parameters:
    show_plots: if True, display the plots; if False, only save them
    plots_folder: folder where plots should be saved
    """
    filename = os.path.basename(file_path)
    
    # Use plots_folder if provided, otherwise use same directory as file
    if plots_folder:
        output_dir = plots_folder
    else:
        output_dir = os.path.dirname(file_path)
    
    base_name = os.path.splitext(filename)[0]
    
    # Calculate Nyquist frequency
    nyquist_freq = sampling_rate / 2
    freq_mask = frequencies <= nyquist_freq
    
    # Calculate Vpp and peak frequency
    vpp = np.max(selected_voltage) - np.min(selected_voltage)
    peak_idx = np.argmax(amplitude_linear)
    peak_freq = frequencies[peak_idx]
    peak_amplitude_linear = amplitude_linear[peak_idx]
    peak_amplitude_db = amplitude_db[peak_idx]
    
    # ==== Create LINEAR scale version ====
    fig_linear, axes_linear = plt.subplots(3, 1, figsize=(18, 16))
    
    # Plot 1: Full waveform (same for both)
    ax1 = axes_linear[0]
    ax1.plot(time_data * 1e6, voltage_data, 'b-', alpha=0.5, linewidth=2, label='Full waveform')
    ax1.plot(selected_time * 1e6, selected_voltage, 'r-', linewidth=3, label='Analysis window (30-90 s)')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Voltage (V)')
    ax1.set_title(f'Full Waveform - {filename}')
    #ax1.grid(False)
    ax1.legend()
    ax1.axvspan(30, 90, alpha=0.1, color='red')
    ax1.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax1.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    # Plot 2: Selected window (same for both)
    ax2 = axes_linear[1]
    ax2.plot(selected_time * 1e6, selected_voltage, 'r-', linewidth=2.5)
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Voltage (V)')
    ax2.set_title(f'Selected Window: 30-90 s (Sampling Rate: {sampling_rate/1e6:.1f} MHz)')
    ax2.grid(False)
    ax2.set_xlim(30, 90)
    ax2.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax2.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    # Plot 3: FFT Linear Scale
    ax3 = axes_linear[2]
    ax3.plot(frequencies[freq_mask] / 1e6, amplitude_linear[freq_mask], 'g-', linewidth=2.5)
    ax3.set_xlabel('Frequency (MHz)')
    ax3.set_ylabel('Amplitude (V)')
    ax3.set_title('FFT Amplitude Spectrum - LINEAR Scale (30-90 s window)')
    ax3.set_xlim(0, nyquist_freq / 1e6)
    ax3.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax3.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    plt.tight_layout()
    
    # Save linear version
    linear_path = os.path.join(output_dir, f'{base_name}_FFT_analysis_linear.png')
    fig_linear.savefig(linear_path, dpi=300, bbox_inches='tight')
    print(f"  Linear scale plot saved: {linear_path}")
    
    # ==== Create LOGARITHMIC (dB) scale version ====
    fig_db, axes_db = plt.subplots(3, 1, figsize=(18, 16))
    
    # Plot 1: Full waveform (same)
    ax1_db = axes_db[0]
    ax1_db.plot(time_data * 1e6, voltage_data, 'b-', alpha=0.5, linewidth=2, label='Full waveform')
    ax1_db.plot(selected_time * 1e6, selected_voltage, 'r-', linewidth=3, label='Analysis window (30-90 s)')
    ax1_db.set_xlabel('Time (s)')
    ax1_db.set_ylabel('Voltage (V)')
    ax1_db.set_title(f'Full Waveform - {filename}')
    ax1_db.grid(False)
    ax1_db.legend()
    ax1_db.axvspan(30, 90, alpha=0.1, color='red')
    ax1_db.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax1_db.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    # Plot 2: Selected window (same)
    ax2_db = axes_db[1]
    ax2_db.plot(selected_time * 1e6, selected_voltage, 'r-', linewidth=2.5)
    ax2_db.set_xlabel('Time (s)')
    ax2_db.set_ylabel('Voltage (V)')
    ax2_db.set_title(f'Selected Window: 30-90 s (Sampling Rate: {sampling_rate/1e6:.1f} MHz)')
    ax2_db.grid(False)
    ax2_db.set_xlim(30, 90)
    ax2_db.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax2_db.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    # Plot 3: FFT dB Scale
    ax3_db = axes_db[2]
    ax3_db.plot(frequencies[freq_mask] / 1e6, amplitude_db[freq_mask], 'g-', linewidth=2.5)
    ax3_db.set_xlabel('Frequency (MHz)')
    ax3_db.set_ylabel('Amplitude (dB)')
    ax3_db.set_title('FFT Amplitude Spectrum - LOGARITHMIC Scale (30-90 s window)')
    ax3_db.set_xlim(0, nyquist_freq / 1e6)
    ax3_db.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax3_db.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    plt.tight_layout()
    
    # Save dB version
    db_path = os.path.join(output_dir, f'{base_name}_FFT_analysis_dB.png')
    fig_db.savefig(db_path, dpi=300, bbox_inches='tight')
    print(f"  dB scale plot saved: {db_path}")
    
    # Also save individual FFT plots for both scales
    save_individual_fft_plots(frequencies, amplitude_linear, amplitude_db, 
                             nyquist_freq, output_dir, base_name)
    
    # Save time domain plots (only need one version since they're the same)
    save_time_domain_plots(time_data, voltage_data, selected_time, selected_voltage,
                          output_dir, base_name, filename, sampling_rate)
    
    # Show both plots if requested (they'll be displayed until user closes them)
    if show_plots:
        plt.show()
    else:
        # Close figures to free memory if not showing
        plt.close(fig_linear)
        plt.close(fig_db)
    
    return vpp, peak_freq, peak_amplitude_linear, peak_amplitude_db

def save_individual_fft_plots(frequencies, amplitude_linear, amplitude_db, 
                              nyquist_freq, output_dir, base_name):
    """Save individual FFT plots in both scales"""
    freq_mask = frequencies <= nyquist_freq
    
    # Linear scale FFT
    fig_fft_linear, ax_fft_linear = plt.subplots(1, 1, figsize=(14, 8))
    ax_fft_linear.plot(frequencies[freq_mask] / 1e6, amplitude_linear[freq_mask], 'g-', linewidth=2.5)
    ax_fft_linear.set_xlabel('Frequency (MHz)')
    ax_fft_linear.set_ylabel('Amplitude (V)')
    ax_fft_linear.set_title('FFT Amplitude Spectrum - LINEAR Scale (30-90 s window)')
    ax_fft_linear.set_xlim(0, nyquist_freq / 1e6)
    ax_fft_linear.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_fft_linear.yaxis.set_minor_locator(AutoMinorLocator(2))
    fft_linear_path = os.path.join(output_dir, f'{base_name}_FFT_linear.png')
    fig_fft_linear.savefig(fft_linear_path, dpi=300, bbox_inches='tight')
    plt.close(fig_fft_linear)
    print(f"  FFT linear plot saved: {fft_linear_path}")
    
    # dB scale FFT
    fig_fft_db, ax_fft_db = plt.subplots(1, 1, figsize=(14, 8))
    ax_fft_db.plot(frequencies[freq_mask] / 1e6, amplitude_db[freq_mask], 'g-', linewidth=2.5)
    ax_fft_db.set_xlabel('Frequency (MHz)')
    ax_fft_db.set_ylabel('Amplitude (dB)')
    ax_fft_db.set_title('FFT Amplitude Spectrum - LOGARITHMIC Scale (30-90 s window)')
    ax_fft_db.set_xlim(0, nyquist_freq / 1e6)
    ax_fft_db.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_fft_db.yaxis.set_minor_locator(AutoMinorLocator(2))
    fft_db_path = os.path.join(output_dir, f'{base_name}_FFT_dB.png')
    fig_fft_db.savefig(fft_db_path, dpi=300, bbox_inches='tight')
    plt.close(fig_fft_db)
    print(f"  FFT dB plot saved: {fft_db_path}")

def save_time_domain_plots(time_data, voltage_data, selected_time, selected_voltage,
                           output_dir, base_name, filename, sampling_rate):
    """Save time domain plots (full waveform and selected window)"""
    # Full waveform
    fig_full, ax_full = plt.subplots(1, 1, figsize=(14, 8))
    ax_full.plot(time_data * 1e6, voltage_data, 'b-', alpha=0.5, linewidth=2, label='Full waveform')
    ax_full.plot(selected_time * 1e6, selected_voltage, 'r-', linewidth=3, label='Analysis window (30-90 s)')
    ax_full.set_xlabel('Time (s)')
    ax_full.set_ylabel('Voltage (V)')
    ax_full.set_title(f'Full Waveform - {filename}')
    ax_full.grid(False)
    ax_full.legend()
    ax_full.axvspan(30, 90, alpha=0.1, color='red')
    ax_full.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_full.yaxis.set_minor_locator(AutoMinorLocator(2))
    full_path = os.path.join(output_dir, f'{base_name}_full_waveform.png')
    fig_full.savefig(full_path, dpi=300, bbox_inches='tight')
    plt.close(fig_full)
    print(f"  Full waveform saved: {full_path}")
    
    # Selected window
    fig_window, ax_window = plt.subplots(1, 1, figsize=(14, 8))
    ax_window.plot(selected_time * 1e6, selected_voltage, 'r-', linewidth=2.5)
    ax_window.set_xlabel('Time (s)')
    ax_window.set_ylabel('Voltage (V)')
    ax_window.set_title(f'Selected Window: 30-90 s (Sampling Rate: {sampling_rate/1e6:.1f} MHz)')
    ax_window.grid(False)
    ax_window.set_xlim(30, 90)
    ax_window.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_window.yaxis.set_minor_locator(AutoMinorLocator(2))
    window_path = os.path.join(output_dir, f'{base_name}_selected_window.png')
    fig_window.savefig(window_path, dpi=300, bbox_inches='tight')
    plt.close(fig_window)
    print(f"  Selected window saved: {window_path}")

def plot_comparison_both_scales(rpmi_data, sample_data, plots_folder):
    """
    Plot RPMI vs Sample comparison in both linear and dB scales
    Save each subplot as an individual file
    
    Parameters:
    plots_folder: folder where comparison plots should be saved
    """
    
    # Calculate common parameters
    nyquist_freq = min(rpmi_data['sampling_rate'], sample_data['sampling_rate']) / 2
    rpmi_freq_mask = rpmi_data['frequencies'] <= nyquist_freq
    sample_freq_mask = sample_data['frequencies'] <= nyquist_freq
    
    # ========== Save Individual Plots ==========
    
    # 1. Time Domain Comparison (same for both scales)
    fig_time, ax_time = plt.subplots(1, 1, figsize=(14, 8))
    # SAMPLE FIRST for legend order
    ax_time.plot(sample_data['selected_time'] * 1e6, sample_data['selected_voltage'], 
                 'r-', linewidth=2, label='Sample', alpha=0.7)
    ax_time.plot(rpmi_data['selected_time'] * 1e6, rpmi_data['selected_voltage'], 
                 'b-', linewidth=2, label='RPMI', alpha=0.7)
    ax_time.set_xlabel('Time (s)')
    ax_time.set_ylabel('Voltage (V)')
    ax_time.set_title('Time Domain Comparison (30-90 s window)')
    # ax_time.grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    ax_time.legend()
    ax_time.set_xlim(30, 90)
    ax_time.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_time.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    time_comparison_path = os.path.join(plots_folder, 'comparison_time_domain.png')
    fig_time.savefig(time_comparison_path, dpi=300, bbox_inches='tight')
    plt.close(fig_time)
    print(f"  Time domain comparison saved: {time_comparison_path}")
    
    # 2. FFT Comparison - Linear Scale
    fig_fft_linear, ax_fft_linear = plt.subplots(1, 1, figsize=(14, 8))
    # SAMPLE FIRST for legend order
    ax_fft_linear.plot(sample_data['frequencies'][sample_freq_mask] / 1e6, 
                       sample_data['amplitude_linear'][sample_freq_mask], 
                       'r-', linewidth=2, label='Sample', alpha=0.7)
    ax_fft_linear.plot(rpmi_data['frequencies'][rpmi_freq_mask] / 1e6, 
                       rpmi_data['amplitude_linear'][rpmi_freq_mask], 
                       'b-', linewidth=2, label='RPMI', alpha=0.7)
    ax_fft_linear.set_xlabel('Frequency (MHz)')
    ax_fft_linear.set_ylabel('Amplitude (V)')
    ax_fft_linear.set_title('FFT Comparison - Linear Scale')
    # ax_fft_linear.grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    ax_fft_linear.legend()
    ax_fft_linear.set_xlim(0, nyquist_freq / 1e6)
    ax_fft_linear.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_fft_linear.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    fft_linear_path = os.path.join(plots_folder, 'comparison_FFT_linear.png')
    fig_fft_linear.savefig(fft_linear_path, dpi=300, bbox_inches='tight')
    plt.close(fig_fft_linear)
    print(f"  FFT linear comparison saved: {fft_linear_path}")
    
    # 3. FFT Comparison - dB Scale
    fig_fft_db, ax_fft_db = plt.subplots(1, 1, figsize=(14, 8))
    # SAMPLE FIRST for legend order
    ax_fft_db.plot(sample_data['frequencies'][sample_freq_mask] / 1e6, 
                   sample_data['amplitude_db'][sample_freq_mask], 
                   'r-', linewidth=2, label='Sample', alpha=0.7)
    ax_fft_db.plot(rpmi_data['frequencies'][rpmi_freq_mask] / 1e6, 
                   rpmi_data['amplitude_db'][rpmi_freq_mask], 
                   'b-', linewidth=2, label='RPMI', alpha=0.7)
    ax_fft_db.set_xlabel('Frequency (MHz)')
    ax_fft_db.set_ylabel('Amplitude (dB)')
    ax_fft_db.set_title('FFT Comparison - Logarithmic Scale')
    # ax_fft_db.grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    ax_fft_db.legend()
    ax_fft_db.set_xlim(0, nyquist_freq / 1e6)
    ax_fft_db.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_fft_db.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    fft_db_path = os.path.join(plots_folder, 'comparison_FFT_dB.png')
    fig_fft_db.savefig(fft_db_path, dpi=300, bbox_inches='tight')
    plt.close(fig_fft_db)
    print(f"  FFT dB comparison saved: {fft_db_path}")
    
    # 4. Difference Plot - Linear Scale
    fig_diff_linear, ax_diff_linear = plt.subplots(1, 1, figsize=(14, 8))
    common_freq = np.linspace(0, nyquist_freq, 1000)
    rpmi_interp = np.interp(common_freq, 
                           rpmi_data['frequencies'][rpmi_freq_mask], 
                           rpmi_data['amplitude_linear'][rpmi_freq_mask])
    sample_interp = np.interp(common_freq, 
                             sample_data['frequencies'][sample_freq_mask], 
                             sample_data['amplitude_linear'][sample_freq_mask])
    
    difference = sample_interp - rpmi_interp
    ax_diff_linear.plot(common_freq / 1e6, difference, 'g-', linewidth=2)
    ax_diff_linear.set_xlabel('Frequency (MHz)')
    ax_diff_linear.set_ylabel('Amplitude Difference (V)')
    ax_diff_linear.set_title('Sample - RPMI Difference (Linear Scale)')
    # ax_diff_linear.grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    ax_diff_linear.axhline(y=0, color='k', linestyle='--', alpha=0.3)
    ax_diff_linear.set_xlim(0, nyquist_freq / 1e6)
    ax_diff_linear.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_diff_linear.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    diff_linear_path = os.path.join(plots_folder, 'comparison_difference_linear.png')
    fig_diff_linear.savefig(diff_linear_path, dpi=300, bbox_inches='tight')
    plt.close(fig_diff_linear)
    print(f"  Difference plot (linear) saved: {diff_linear_path}")
    
    # 5. Difference Plot - dB Scale
    fig_diff_db, ax_diff_db = plt.subplots(1, 1, figsize=(14, 8))
    rpmi_interp_db = np.interp(common_freq, 
                               rpmi_data['frequencies'][rpmi_freq_mask], 
                               rpmi_data['amplitude_db'][rpmi_freq_mask])
    sample_interp_db = np.interp(common_freq, 
                                 sample_data['frequencies'][sample_freq_mask], 
                                 sample_data['amplitude_db'][sample_freq_mask])
    
    difference_db = sample_interp_db - rpmi_interp_db
    ax_diff_db.plot(common_freq / 1e6, difference_db, 'g-', linewidth=2)
    ax_diff_db.set_xlabel('Frequency (MHz)')
    ax_diff_db.set_ylabel('Amplitude Difference (dB)')
    ax_diff_db.set_title('Sample - RPMI Difference (dB Scale)')
    # ax_diff_db.grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    ax_diff_db.axhline(y=0, color='k', linestyle='--', alpha=0.3)
    ax_diff_db.set_xlim(0, nyquist_freq / 1e6)
    ax_diff_db.xaxis.set_minor_locator(AutoMinorLocator(2))
    ax_diff_db.yaxis.set_minor_locator(AutoMinorLocator(2))
    
    diff_db_path = os.path.join(plots_folder, 'comparison_difference_dB.png')
    fig_diff_db.savefig(diff_db_path, dpi=300, bbox_inches='tight')
    plt.close(fig_diff_db)
    print(f"  Difference plot (dB) saved: {diff_db_path}")
    
    # 6. Statistics Comparison - Save as text file and as figure
    stats_text = f"""COMPARISON STATISTICS
{'='*50}

Sample:
  File: {os.path.basename(sample_data['file_path'])}
  Vpp: {sample_data['vpp']:.4f} V
  Peak Frequency: {sample_data['peak_freq']/1e6:.3f} MHz
  Peak Amplitude (Linear): {sample_data['peak_amplitude_linear']:.4f} V
  Peak Amplitude (dB): {sample_data['peak_amplitude_db']:.2f} dB
  Sampling Rate: {sample_data['sampling_rate']/1e6:.1f} MHz

RPMI:
  File: {os.path.basename(rpmi_data['file_path'])}
  Vpp: {rpmi_data['vpp']:.4f} V
  Peak Frequency: {rpmi_data['peak_freq']/1e6:.3f} MHz
  Peak Amplitude (Linear): {rpmi_data['peak_amplitude_linear']:.4f} V
  Peak Amplitude (dB): {rpmi_data['peak_amplitude_db']:.2f} dB
  Sampling Rate: {rpmi_data['sampling_rate']/1e6:.1f} MHz

Differences:
  Delta Vpp: {sample_data['vpp'] - rpmi_data['vpp']:.4f} V
  Delta Peak Freq: {(sample_data['peak_freq'] - rpmi_data['peak_freq'])/1e6:.3f} MHz
  Delta Peak Amplitude (Linear): {sample_data['peak_amplitude_linear'] - rpmi_data['peak_amplitude_linear']:.4f} V
  Delta Peak Amplitude (dB): {sample_data['peak_amplitude_db'] - rpmi_data['peak_amplitude_db']:.2f} dB
"""
    
    # Save statistics as text file
    stats_text_path = os.path.join(plots_folder, 'comparison_statistics.txt')
    with open(stats_text_path, 'w', encoding='utf-8') as f:
        f.write(stats_text)
    print(f"  Statistics saved: {stats_text_path}")
    
    # Also save statistics as a figure
    fig_stats, ax_stats = plt.subplots(1, 1, figsize=(12, 10))
    ax_stats.axis('off')
    ax_stats.text(0.05, 0.5, stats_text, fontsize=14, family='monospace', 
                  verticalalignment='center', transform=ax_stats.transAxes)
    
    stats_fig_path = os.path.join(plots_folder, 'comparison_statistics.png')
    fig_stats.savefig(stats_fig_path, dpi=300, bbox_inches='tight')
    plt.close(fig_stats)
    print(f"  Statistics figure saved: {stats_fig_path}")
    
    # ========== Optional: Still create combined plots ==========
    # You can comment out this section if you don't want the combined plots at all
    
    # Create combined plot for LINEAR scale
    fig_linear, axes_linear = plt.subplots(2, 2, figsize=(20, 14))
    fig_linear.suptitle('RPMI vs Sample Comparison - LINEAR Scale', fontsize=26)
    
    # Recreate all subplots in combined figure
    # Time domain - SAMPLE FIRST
    axes_linear[0, 0].plot(sample_data['selected_time'] * 1e6, sample_data['selected_voltage'], 
                          'r-', linewidth=2, label='Sample', alpha=0.7)
    axes_linear[0, 0].plot(rpmi_data['selected_time'] * 1e6, rpmi_data['selected_voltage'], 
                          'b-', linewidth=2, label='RPMI', alpha=0.7)
    axes_linear[0, 0].set_xlabel('Time (s)')
    axes_linear[0, 0].set_ylabel('Voltage (V)')
    axes_linear[0, 0].set_title('Time Domain Comparison (30-90 s window)')
    # axes_linear[0, 0].grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    axes_linear[0, 0].legend()
    axes_linear[0, 0].set_xlim(30, 90)
    
    # FFT Linear - SAMPLE FIRST
    axes_linear[0, 1].plot(sample_data['frequencies'][sample_freq_mask] / 1e6, 
                          sample_data['amplitude_linear'][sample_freq_mask], 
                          'r-', linewidth=2, label='Sample', alpha=0.7)
    axes_linear[0, 1].plot(rpmi_data['frequencies'][rpmi_freq_mask] / 1e6, 
                          rpmi_data['amplitude_linear'][rpmi_freq_mask], 
                          'b-', linewidth=2, label='RPMI', alpha=0.7)
    axes_linear[0, 1].set_xlabel('Frequency (MHz)')
    axes_linear[0, 1].set_ylabel('Amplitude (V)')
    axes_linear[0, 1].set_title('FFT Comparison - Linear Scale')
    # axes_linear[0, 1].grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    axes_linear[0, 1].legend()
    axes_linear[0, 1].set_xlim(0, nyquist_freq / 1e6)
    
    # Difference
    axes_linear[1, 0].plot(common_freq / 1e6, difference, 'g-', linewidth=2)
    axes_linear[1, 0].set_xlabel('Frequency (MHz)')
    axes_linear[1, 0].set_ylabel('Amplitude Difference (V)')
    axes_linear[1, 0].set_title('Sample - RPMI Difference (Linear Scale)')
    # axes_linear[1, 0].grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    axes_linear[1, 0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes_linear[1, 0].set_xlim(0, nyquist_freq / 1e6)
    
    # Statistics
    axes_linear[1, 1].axis('off')
    axes_linear[1, 1].text(0.1, 0.5, stats_text, fontsize=16, family='monospace', 
                          verticalalignment='center', transform=axes_linear[1, 1].transAxes)
    
    plt.tight_layout()
    
    # Save combined linear version
    combined_linear_path = os.path.join(plots_folder, 'RPMI_vs_Sample_comparison_linear_combined.png')
    fig_linear.savefig(combined_linear_path, dpi=300, bbox_inches='tight')
    print(f"  Combined linear plot saved: {combined_linear_path}")
    
    # Create combined plot for dB scale
    fig_db, axes_db = plt.subplots(2, 2, figsize=(20, 14))
    fig_db.suptitle('RPMI vs Sample Comparison - LOGARITHMIC Scale', fontsize=26)
    
    # Time domain - SAMPLE FIRST
    axes_db[0, 0].plot(sample_data['selected_time'] * 1e6, sample_data['selected_voltage'], 
                       'r-', linewidth=2, label='Sample', alpha=0.7)
    axes_db[0, 0].plot(rpmi_data['selected_time'] * 1e6, rpmi_data['selected_voltage'], 
                       'b-', linewidth=2, label='RPMI', alpha=0.7)
    axes_db[0, 0].set_xlabel('Time (s)')
    axes_db[0, 0].set_ylabel('Voltage (V)')
    axes_db[0, 0].set_title('Time Domain Comparison (30-90 s window)')
    # axes_db[0, 0].grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    axes_db[0, 0].legend()
    axes_db[0, 0].set_xlim(30, 90)
    
    # FFT dB - SAMPLE FIRST
    axes_db[0, 1].plot(sample_data['frequencies'][sample_freq_mask] / 1e6, 
                       sample_data['amplitude_db'][sample_freq_mask], 
                       'r-', linewidth=2, label='Sample', alpha=0.7)
    axes_db[0, 1].plot(rpmi_data['frequencies'][rpmi_freq_mask] / 1e6, 
                       rpmi_data['amplitude_db'][rpmi_freq_mask], 
                       'b-', linewidth=2, label='RPMI', alpha=0.7)
    axes_db[0, 1].set_xlabel('Frequency (MHz)')
    axes_db[0, 1].set_ylabel('Amplitude (dB)')
    axes_db[0, 1].set_title('FFT Comparison - Logarithmic Scale')
    # axes_db[0, 1].grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    axes_db[0, 1].legend()
    axes_db[0, 1].set_xlim(0, nyquist_freq / 1e6)
    
    # Difference dB
    axes_db[1, 0].plot(common_freq / 1e6, difference_db, 'g-', linewidth=2)
    axes_db[1, 0].set_xlabel('Frequency (MHz)')
    axes_db[1, 0].set_ylabel('Amplitude Difference (dB)')
    axes_db[1, 0].set_title('Sample - RPMI Difference (dB Scale)')
    # axes_db[1, 0].grid(True, alpha=0.4)  # Remove or set to False if no grid wanted
    axes_db[1, 0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes_db[1, 0].set_xlim(0, nyquist_freq / 1e6)
    
    axes_db[1, 1].axis('off')
    axes_db[1, 1].text(0.1, 0.5, stats_text, fontsize=16, family='monospace', 
                       verticalalignment='center', transform=axes_db[1, 1].transAxes)
    
    plt.tight_layout()
    
    # Save combined dB version
    combined_db_path = os.path.join(plots_folder, 'RPMI_vs_Sample_comparison_dB_combined.png')
    fig_db.savefig(combined_db_path, dpi=300, bbox_inches='tight')
    print(f"  Combined dB plot saved: {combined_db_path}")
    
    # Show plots if desired
    plt.show()
    
    print(f"\n{'='*60}")
    print("Individual comparison plots saved:")
    print(f"   comparison_time_domain.png")
    print(f"   comparison_FFT_linear.png")
    print(f"   comparison_FFT_dB.png")
    print(f"   comparison_difference_linear.png")
    print(f"   comparison_difference_dB.png")
    print(f"   comparison_statistics.txt")
    print(f"   comparison_statistics.png")
    print(f"\nCombined comparison plots saved:")
    print(f"   RPMI_vs_Sample_comparison_linear_combined.png")
    print(f"   RPMI_vs_Sample_comparison_dB_combined.png")
    print(f"{'='*60}")

def main():
    """
    Main function to process files
    """
    # Select files or folder
    file_paths, base_folder = select_files_or_folder()
    
    if not file_paths:
        print("No files selected. Exiting.")
        return
    
    # Create a subfolder for all plots in the originally selected folder
    # Use base_folder which is the folder the user actually selected
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    plots_folder = os.path.join(base_folder, f"FFT_Analysis_Plots_{timestamp}")
    
    # Create the plots folder
    os.makedirs(plots_folder, exist_ok=True)
    print(f"\nCreated plots folder: {plots_folder}")
    
    # If multiple files, ask about plot display
    show_plots = True
    if len(file_paths) > 2:
        print(f"\nProcessing {len(file_paths)} files.")
        show_plots_input = input("Display all plots during processing? (y/n): ").lower()
        show_plots = show_plots_input == 'y'
        if not show_plots:
            print("Plots will only be saved to disk (not displayed)")
        else:
            print("Plots will be displayed (may open many windows)")
    
    print("\n" + "="*60)
    print("Processing files - Generating BOTH Linear and dB scale plots")
    print(f"Selected folder: {base_folder}")
    print(f"Plots will be saved to: {os.path.basename(plots_folder)}")
    print("="*60)
    
    # Process each file
    results = []
    processed_data = []  # Store processed data for potential comparison
    
    for file_path in file_paths:
        print(f"\nProcessing: {os.path.basename(file_path)}")
        print("-" * 50)
        
        try:
            # Read data
            time_data, voltage_data = read_oscilloscope_csv(file_path)
            
            # Perform FFT analysis on 30-90 s window - get both scales
            frequencies, amplitude_linear, amplitude_db, selected_time, selected_voltage, sampling_rate = \
                perform_fft_analysis(time_data, voltage_data, start_time=30e-6, end_time=90e-6)
            
            # Plot results in both scales
            vpp, peak_freq, peak_amplitude_linear, peak_amplitude_db = \
                plot_results_both_scales(time_data, voltage_data, frequencies, 
                                       amplitude_linear, amplitude_db,
                                       selected_time, selected_voltage, 
                                       file_path, sampling_rate, show_plots, plots_folder)
            
            # Store processed data
            processed_data.append({
                'file_path': file_path,
                'time_data': time_data,
                'voltage_data': voltage_data,
                'frequencies': frequencies,
                'amplitude_linear': amplitude_linear,
                'amplitude_db': amplitude_db,
                'selected_time': selected_time,
                'selected_voltage': selected_voltage,
                'sampling_rate': sampling_rate,
                'vpp': vpp,
                'peak_freq': peak_freq,
                'peak_amplitude_linear': peak_amplitude_linear,
                'peak_amplitude_db': peak_amplitude_db
            })
            
            # Store results for summary
            results.append({
                'File': os.path.basename(file_path),
                'Vpp (V)': f'{vpp:.4f}',
                'Peak Frequency (MHz)': f'{peak_freq/1e6:.3f}',
                'Peak Amplitude (V)': f'{peak_amplitude_linear:.4f}',
                'Peak Amplitude (dB)': f'{peak_amplitude_db:.2f}',
                'Sampling Rate (MHz)': f'{sampling_rate/1e6:.1f}'
            })
            
            print(f"  Vpp: {vpp:.3f} V")
            print(f"  Peak Frequency: {peak_freq/1e6:.3f} MHz")
            print(f"  Peak Amplitude: {peak_amplitude_linear:.4f} V / {peak_amplitude_db:.2f} dB")
            print(f"  Sampling Rate: {sampling_rate/1e6:.1f} MHz")
            print(f"  Generated both Linear and dB scale plots")
            
        except Exception as e:
            print(f"  Error processing file: {e}")
            continue
    
    # Display summary if multiple files
    if len(results) > 1:
        print("\n" + "=" * 60)
        print("SUMMARY OF ALL FILES")
        print("=" * 60)
        results_df = pd.DataFrame(results)
        print(results_df.to_string(index=False))
        
        # Ask if user wants to plot files together for comparison
        if len(processed_data) >= 2:
            plot_together_input = input("\nWould you like to create a comparison plot? (y/n): ").lower()
            plot_together = plot_together_input == 'y'
            
            if plot_together:
                # Show available files with numbers
                print("\n" + "="*60)
                print("Available files for comparison:")
                print("="*60)
                for i, data in enumerate(processed_data, 1):
                    print(f"  {i}. {os.path.basename(data['file_path'])}")
                print("="*60)
                
                # Simple console input for RPMI and Sample selection
                while True:
                    try:
                        rpmi_num = input("\nEnter the number for RPMI file: ")
                        rpmi_idx = int(rpmi_num) - 1
                        
                        if 0 <= rpmi_idx < len(processed_data):
                            break
                        else:
                            print(f"Please enter a number between 1 and {len(processed_data)}")
                    except ValueError:
                        print("Please enter a valid number")
                
                while True:
                    try:
                        sample_num = input("Enter the number for Sample file: ")
                        sample_idx = int(sample_num) - 1
                        
                        if 0 <= sample_idx < len(processed_data):
                            break
                        else:
                            print(f"Please enter a number between 1 and {len(processed_data)}")
                    except ValueError:
                        print("Please enter a valid number")
                
                # Get the selected data
                rpmi_data = processed_data[rpmi_idx]
                sample_data = processed_data[sample_idx]
                
                print(f"\nCreating comparison:")
                print(f"  RPMI: {os.path.basename(rpmi_data['file_path'])}")
                print(f"  Sample: {os.path.basename(sample_data['file_path'])}")
                
                # Create the comparison plots
                plot_comparison_both_scales(rpmi_data, sample_data, plots_folder)
        
        # Optional: Save summary to CSV
        save_summary_input = input("\nSave summary to CSV? (y/n): ").lower()
        if save_summary_input == 'y':
            # Save in the plots folder with timestamp
            csv_filename = f"FFT_Analysis_Summary.csv"
            output_path = os.path.join(plots_folder, csv_filename)
            results_df.to_csv(output_path, index=False)
            print(f"Summary saved to: {output_path}")
    
    # Summary of saved plots
    print(f"\n{'='*60}")
    print(f"All plots saved in: {plots_folder}")
    print(f"Generated BOTH Linear and Logarithmic (dB) scale versions")
    print(f"\nFiles saved per CSV:")
    print(f"  === 3-Panel Combined Plots ===")
    print(f"   {'{name}'}_FFT_analysis_linear.png (Linear scale)")
    print(f"   {'{name}'}_FFT_analysis_dB.png (Logarithmic scale)")
    print(f"\n  === Individual Plots ===")
    print(f"   {'{name}'}_full_waveform.png (Full signal)")
    print(f"   {'{name}'}_selected_window.png (30-90 s window)")
    print(f"   {'{name}'}_FFT_linear.png (FFT - Linear scale)")
    print(f"   {'{name}'}_FFT_dB.png (FFT - Logarithmic scale)")
    
    if len(processed_data) >= 2:
        print(f"\n  === Comparison Plots (if created) ===")
        print(f"   RPMI_vs_Sample_comparison_linear.png")
        print(f"   RPMI_vs_Sample_comparison_dB.png")
    print(f"\nPlots folder: {plots_folder}")
    print(f"{'='*60}")

if __name__ == "__main__":
    main()
```
